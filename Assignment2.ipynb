```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
t = np.arange(-50, 50, 0.01)
x1 = 3 * np.cos(0.5 * t) + 2 * np.sin(0.2 * t)
x2 = 2 * np.cos(0.3 * t) + np.sin(0.1 * t)

plt.plot(t, x1, label='x1 (Low Frequency Signal)')
plt.plot(t, x2, label='x2 (Low Frequency Signal)')
plt.title('Low Frequency Signals')
plt.xlabel('Time (t)')
plt.ylabel('Amplitude')
plt.legend()
plt.grid()
plt.show()
```

```python
x1m = x1 * np.cos(10 * t)
x2m = x2 * np.cos(20 * t)

plt.figure()
plt.plot(t, x1m, label='x1m (Modulated with cos(10t))')
plt.plot(t, x2m, label='x2m (Modulated with cos(20t))')
plt.title('Modulated Signals')
plt.xlabel('Time (t)')
plt.ylabel('Amplitude')
plt.legend()
plt.grid()
plt.show()
```

```python
x = x1m + x2m

plt.figure()
plt.plot(t, x, label='x (Sum of Modulated Signals)')
plt.title('Combined Signal')
plt.xlabel('Time (t)')
plt.ylabel('Amplitude')
plt.legend()
plt.grid()
plt.show()
```

```python
bandpass1 = np.cos(10 * t) * np.exp(-t**2 / 100)
bandpass2 = np.cos(20 * t) * np.exp(-t**2 / 100)

x1m_extracted = 0.01 * np.convolve(x, bandpass1, mode='same')
x2m_extracted = 0.01 * np.convolve(x, bandpass2, mode='same')

plt.figure()
plt.plot(t, x1m_extracted, label='x1m_extracted')
plt.plot(t, x2m_extracted, label='x2m_extracted')
plt.title('Extracted Modulated Signals')
plt.xlabel('Time (t)')
plt.ylabel('Amplitude')
plt.legend()
plt.grid()
plt.show()
```

```python
lowpass = np.exp(-t**2 / 100)

x1_reconstructed = 0.01 * np.convolve(x1m_extracted * np.cos(10 * t), lowpass, mode='same')
x2_reconstructed = 0.01 * np.convolve(x2m_extracted * np.cos(20 * t), lowpass, mode='same')

plt.figure()
plt.plot(t, x1, label='Original x1')
plt.plot(t, x1_reconstructed, label='Reconstructed x1')
plt.title('Original vs Reconstructed x1')
plt.xlabel('Time (t)')
plt.ylabel('Amplitude')
plt.legend()
plt.grid()
plt.show()

plt.figure()
plt.plot(t, x2, label='Original x2')
plt.plot(t, x2_reconstructed, label='Reconstructed x2')
plt.title('Original vs Reconstructed x2')
plt.xlabel('Time (t)')
plt.ylabel('Amplitude')
plt.legend()
plt.grid()
plt.show()
```
